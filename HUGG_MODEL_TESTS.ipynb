# MASKED TOKEN GENERATION

```python
from transformers import pipeline

# Step 1: Input sentence
sentence = "Artificial intelligence is revolutionizing industries worldwide."

# Step 2: Split the sentence into words and mask all words except the first one
words = sentence.split()
masked_text = words[0] + " " + " ".join(["[MASK]"] * (len(words) - 1))

# Step 3: Load two models for masked word prediction
mlm_model_1 = pipeline('fill-mask', model='distilbert-base-uncased')  # Model 1: DistilBERT
mlm_model_2 = pipeline('fill-mask', model='bert-base-uncased')        # Model 2: BERT Base

def generate_sequence(model, masked_text, model_name):
    """Generates the full sequence by iteratively replacing [MASK] tokens."""
    print(f"Generation process for {model_name}:\n")
    for step in range(1, len(words)):
        if "[MASK]" not in masked_text:
            break
        # Get the predictions for the current masked text
        output = model(masked_text)

        # Handle nested outputs
        if isinstance(output, list) and len(output) > 0 and isinstance(output[0], list):
            output = output[0]  # Unnest the first list level

        # Check if we have valid predictions
        if isinstance(output, list) and len(output) > 0 and isinstance(output[0], dict):
            predicted_word = output[0].get('token_str', '')  # Extract the top prediction
            if predicted_word == "[MASK]":  # Handle case where no new token is generated
                predicted_word = ""  # Replace mask with an empty string or placeholder

            print(f"Step {step}: Predicted word: '{predicted_word}'")
            print(f"Before replacement: {masked_text}")

            masked_text = masked_text.replace("[MASK]", predicted_word, 1)  # Replace one mask

            print(f"After replacement: {masked_text}\n")
        else:
            print("Unexpected output structure:", output)
            break
    return masked_text

# Step 4: Generate sequence using Model 1
print("Generating text with Model 1 (DistilBERT)...")
final_result_1 = generate_sequence(mlm_model_1, masked_text, "DistilBERT")
print("\nFinal Result with Model 1 (DistilBERT):")
print(final_result_1)

# Step 5: Generate sequence using Model 2
print("\nGenerating text with Model 2 (BERT Base)...")
final_result_2 = generate_sequence(mlm_model_2, masked_text, "BERT Base")
print("\nFinal Result with Model 2 (BERT Base):")
print(final_result_2)

# Step 6: Compare results
print("\nComparison:")
print(f"Original Sentence: {sentence}")
print(f"Reconstructed with Model 1 (DistilBERT): {final_result_1}")
print(f"Reconstructed with Model 2 (BERT Base): {final_result_2}")
```

# MASED TOKEN GENERATION WITH LONGER NON MASKED PART

```python
from transformers import pipeline

# Step 1: Input sentences
text = """
Artificial intelligence is transforming industries worldwide.
It is revolutionizing fields like healthcare and education.
The future of AI promises both opportunities and challenges.
"""

# Step 2: Split the text into sentences and mask the words in the second and third sentences
sentences = text.strip().split(". ")  # Split into individual sentences
sentences = [s.strip() for s in sentences if s]  # Clean up extra spaces and remove empty strings

# Preserve the first sentence and mask the remaining sentences
masked_text = sentences[0] + ". " + " ".join(["[MASK]"] * sum(len(s.split()) for s in sentences[1:]))

# Step 3: Load two models for masked word prediction
mlm_model_1 = pipeline('fill-mask', model='distilbert-base-uncased')  # Model 1: DistilBERT
mlm_model_2 = pipeline('fill-mask', model='bert-base-uncased')        # Model 2: BERT Base

def generate_sequence(model, masked_text, model_name):
    """Generates the full sequence by iteratively replacing [MASK] tokens."""
    print(f"Generation process for {model_name}:\n")
    mask_count = masked_text.count("[MASK]")
    for step in range(1, mask_count + 1):
        if "[MASK]" not in masked_text:
            break
        # Get the predictions for the current masked text
        output = model(masked_text)

        # Handle nested outputs
        if isinstance(output, list) and len(output) > 0 and isinstance(output[0], list):
            output = output[0]  # Unnest the first list level

        # Check if we have valid predictions
        if isinstance(output, list) and len(output) > 0 and isinstance(output[0], dict):
            predicted_word = output[0].get('token_str', '')  # Extract the top prediction
            if predicted_word == "[MASK]":  # Handle case where no new token is generated
                predicted_word = ""  # Replace mask with an empty string or placeholder

            print(f"Step {step}: Predicted word: '{predicted_word}'")
            print(f"Before replacement: {masked_text}")

            masked_text = masked_text.replace("[MASK]", predicted_word, 1)  # Replace one mask

            print(f"After replacement: {masked_text}\n")
        else:
            print("Unexpected output structure:", output)
            break
    return masked_text

# Step 4: Generate sequence using Model 1
print("Generating text with Model 1 (DistilBERT)...")
final_result_1 = generate_sequence(mlm_model_1, masked_text, "DistilBERT")
print("\nFinal Result with Model 1 (DistilBERT):")
print(final_result_1)

# Step 5: Generate sequence using Model 2
print("\nGenerating text with Model 2 (BERT Base)...")
final_result_2 = generate_sequence(mlm_model_2, masked_text, "BERT Base")
print("\nFinal Result with Model 2 (BERT Base):")
print(final_result_2)

# Step 6: Compare results
print("\nComparison:")
print(f"Original Text: {text.strip()}")
print(f"Reconstructed with Model 1 (DistilBERT): {final_result_1}")
print(f"Reconstructed with Model 2 (BERT Base): {final_result_2}")
```

```python
from transformers import SpeechT5Processor, SpeechT5ForTextToSpeech, SpeechT5HifiGan
import torch
import soundfile as sf

# Sentences to convert to voice
sentence_1 = "Artificial intelligence is transforming industries worldwide."
sentence_2 = "It is revolutionizing fields like healthcare and education."

# Model 1: SpeechT5
print("Generating voice with Model 1: SpeechT5...")
# Load the SpeechT5 processor, model, and vocoder
processor_t5 = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
model_t5 = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")
vocoder_t5 = SpeechT5HifiGan.from_pretrained("microsoft/speecht5_hifigan")

# Generate voice for sentence_1 using SpeechT5
inputs_t5 = processor_t5(text=sentence_1, return_tensors="pt")
speaker_embedding = torch.zeros((1, 512))  # Dummy speaker embedding
speech_t5 = model_t5.generate_speech(inputs_t5["input_ids"], speaker_embedding, vocoder=vocoder_t5)
sf.write("speech_t5_sentence1.wav", speech_t5.numpy(), 16000)

# Generate voice for sentence_2 using SpeechT5
inputs_t5_2 = processor_t5(text=sentence_2, return_tensors="pt")
speech_t5_2 = model_t5.generate_speech(inputs_t5_2["input_ids"], speaker_embedding, vocoder=vocoder_t5)
sf.write("speech_t5_sentence2.wav", speech_t5_2.numpy(), 16000)

print("Speech generated for Model 1: SpeechT5.")

# Model 2: Bark (or replace with another TTS model)
print("Generating voice with Model 2: Replace with another TTS model...")
# Placeholder: Replace with other TTS model like Bark, Tacotron, or your choice

# Example: Using the same SpeechT5 for demonstration
# Generate voice for sentence_1 using Model 2 (placeholder)
speech_model2_sentence1 = model_t5.generate_speech(inputs_t5["input_ids"], speaker_embedding, vocoder=vocoder_t5)
sf.write("speech_model2_sentence1.wav", speech_model2_sentence1.numpy(), 16000)

# Generate voice for sentence_2 using Model 2 (placeholder)
speech_model2_sentence2 = model_t5.generate_speech(inputs_t5_2["input_ids"], speaker_embedding, vocoder=vocoder_t5)
sf.write("speech_model2_sentence2.wav", speech_model2_sentence2.numpy(), 16000)

print("Speech generated for Model 2.")

# Play the generated files
from IPython.display import Audio
print("\nPlaying audio for Model 1, Sentence 1:")
display(Audio("speech_t5_sentence1.wav", rate=16000))

print("\nPlaying audio for Model 1, Sentence 2:")
display(Audio("speech_t5_sentence2.wav", rate=16000))

print("\nPlaying audio for Model 2, Sentence 1:")
display(Audio("speech_model2_sentence1.wav", rate=16000))

print("\nPlaying audio for Model 2, Sentence 2:")
display(Audio("speech_model2_sentence2.wav", rate=16000))
```

# SENTIMENT ANALYSIS

```python
# Install necessary libraries (uncomment if not already installed)
# !pip install transformers

from transformers import pipeline

# Step 1: Input examples
examples = [
    "I absolutely love the new design of this product!",
    "The experience was terrible and I would not recommend it.",
    "The service was acceptable, but nothing extraordinary."
]

# Step 2: Model 1: DistilBERT Sentiment Analysis
print("Performing sentiment analysis with Model 1: DistilBERT (default Hugging Face sentiment model)...")
sentiment_model_1 = pipeline("sentiment-analysis")  # Default is DistilBERT
results_model_1 = [sentiment_model_1(example)[0] for example in examples]

# Display results for Model 1
for i, result in enumerate(results_model_1):
    print(f"Example {i+1}:")
    print(f"Text: {examples[i]}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.2f}")
    print()

# Step 3: Model 2: BERT-based Sentiment Analysis (nlptown/bert-base-multilingual-uncased-sentiment)
print("\nPerforming sentiment analysis with Model 2: BERT (nlptown/bert-base-multilingual-uncased-sentiment)...")
sentiment_model_2 = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")
results_model_2 = [sentiment_model_2(example)[0] for example in examples]

# Display results for Model 2
for i, result in enumerate(results_model_2):
    print(f"Example {i+1}:")
    print(f"Text: {examples[i]}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.2f}")
    print()

# Step 4: Compare results
print("\nComparison of Sentiments:")
for i in range(len(examples)):
    print(f"Example {i+1}:")
    print(f"Text: {examples[i]}")
    print(f"Model 1 Sentiment: {results_model_1[i]['label']} (Confidence: {results_model_1[i]['score']:.2f})")
    print(f"Model 2 Sentiment: {results_model_2[i]['label']} (Confidence: {results_model_2[i]['score']:.2f})")
    print()
```
